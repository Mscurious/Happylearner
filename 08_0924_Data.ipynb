```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.backends.backend_pdf import PdfPages

# Load the files
tutor_com_df = pd.read_excel('/content/sample_data/Tutor_com_Report.xlsx')
dallas_college_df = pd.read_excel('/content/sample_data/Dallas_College_Report.xlsx')

# Function to plot graphs for a given dataset and save to a PDF
def plot_data(df, institution_name, date_format):
    # Create a PDF file to save the plots
    pdf_filename = f'{institution_name}_report.pdf'
    with PdfPages(pdf_filename) as pdf:
        # Convert 'Session Start Time' to datetime with specified format
        df['Session Start Time'] = pd.to_datetime(df['Session Start Time'], format=date_format, errors='coerce')

        # Split into date and time columns
        df['Session Date'] = df['Session Start Time'].dt.date
        df['Session Time'] = df['Session Start Time'].dt.time
        df['Session Hour'] = df['Session Start Time'].dt.hour

        # 1. Total usage by hours for each subject
        subject_usage_hours = df.groupby('Subject')['Length'].sum().reset_index()

        # Plotting subject usage by hours
        plt.figure(figsize=(10,6))
        sns.barplot(x='Subject', y='Length', data=subject_usage_hours)
        plt.title(f'Total Usage by Hours for Each Subject - {institution_name}')
        plt.xlabel('Subject')
        plt.ylabel('Total Hours Used')
        plt.xticks(rotation=90)
        plt.tight_layout()
        pdf.savefig()  # Save current figure to PDF
        plt.close()  # Close the plot

        # 2. Wait time vs subjects
        wait_time_vs_subjects = df.groupby('Subject')['Wait Time'].mean().reset_index()

        # Plotting wait time vs subjects
        plt.figure(figsize=(10,6))
        sns.barplot(x='Subject', y='Wait Time', data=wait_time_vs_subjects)
        plt.title(f'Average Wait Time for Each Subject - {institution_name}')
        plt.xlabel('Subject')
        plt.ylabel('Wait Time (minutes)')
        plt.xticks(rotation=90)
        plt.tight_layout()
        pdf.savefig()  # Save current figure to PDF
        plt.close()  # Close the plot

        # 3. Number of sessions per subject
        sessions_per_subject = df.groupby('Subject')['Session Start Time'].count().reset_index()
        sessions_per_subject.columns = ['Subject', 'Number of Sessions']

        # Plotting number of sessions per subject
        plt.figure(figsize=(10,6))
        sns.barplot(x='Subject', y='Number of Sessions', data=sessions_per_subject)
        plt.title(f'Number of Sessions per Subject - {institution_name}')
        plt.xlabel('Subject')
        plt.ylabel('Number of Sessions')
        plt.xticks(rotation=90)
        plt.tight_layout()
        pdf.savefig()  # Save current figure to PDF
        plt.close()  # Close the plot

        # 4. Usage hours for each subject
        plt.figure(figsize=(10,6))
        sns.barplot(x='Subject', y='Length', data=df)
        plt.title(f'Usage Hours for Each Subject - {institution_name}')
        plt.xlabel('Subject')
        plt.ylabel('Hours Used')
        plt.xticks(rotation=90)
        plt.tight_layout()
        pdf.savefig()  # Save current figure to PDF
        plt.close()  # Close the plot

        # 5. Tutors and their session lengths
        tutor_session_length = df.groupby('Tutor Display Name')['Length'].sum().reset_index()

        # Plotting tutor session lengths
        plt.figure(figsize=(10,6))
        sns.barplot(x='Tutor Display Name', y='Length', data=tutor_session_length)
        plt.title(f'Total Session Length by Tutor - {institution_name}')
        plt.xlabel('Tutor')
        plt.ylabel('Total Hours')
        plt.xticks(rotation=90)
        plt.tight_layout()
        pdf.savefig()  # Save current figure to PDF
        plt.close()  # Close the plot

    print(f"PDF saved as {pdf_filename}")

# Define the date format (modify according to your dataset)
date_format = '%Y-%m-%d %H:%M:%S'  # Example format; change this to match your actual data format

# Plot for Dallas College
plot_data(dallas_college_df, "Dallas_College", date_format)

# Plot for Tutor.com
plot_data(tutor_com_df, "Tutor_com", date_format)
```

```python
# Function to calculate total session length per tutor and plot
def plot_total_session_length(df, institution_name):
    # Calculate total session length for each tutor
    tutor_session_length = df.groupby('Tutor Display Name')['Length'].sum().reset_index()

    # Sort by session length in descending order for better visualization
    tutor_session_length = tutor_session_length.sort_values(by='Length', ascending=False)

    # Plotting the total session length for each tutor
    plt.figure(figsize=(12,6))
    sns.barplot(x='Tutor Display Name', y='Length', data=tutor_session_length)
    plt.title(f'Total Session Length by Tutor - {institution_name}')
    plt.xlabel('Tutor')
    plt.ylabel('Total Session Length (Hours)')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

# Plot for Dallas College
plot_total_session_length(dallas_college_df, "Dallas College")

# Plot for Tutor.com
plot_total_session_length(tutor_com_df, "Tutor.com")
```
